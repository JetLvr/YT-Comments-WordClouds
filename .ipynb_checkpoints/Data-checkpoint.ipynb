```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud, ImageColorGenerator

pd.set_option('max_colwidth', 400)
```

```python
df["text"] = df.comment.astype(str)
```

```python
df = df.drop(12258)
df.shape[0]
```

```python
# list just one of each name
names = [str(i) for i in df.author.unique()]
len(names),df.shape[0] 
```

```python
# sort by descending name length
sted_names = sorted(names, key=len, reverse=True)

# remove empty names
names = [i for i in sted_names if not i in [""," "]] 

L = []
for c in df.text:
    # search for all names in the comment
    for n in names:         
        n = "@"+n       
        if n in c: 
            # remove the "@+name" if it is in the comment
            a = c.replace(n,"") 
            break
        else:
            a = c
            continue
    L.append(a)
    
df.shape[0],len(L)
```

```python
df["text"] = L
```

```python
# identify rows where text is entirely composed of "white space" characters
df.loc[df.text.str.fullmatch(r"\s*")].head() 
```

```python
df.loc[df.text.str.fullmatch(r"\s*")].shape[0]
```

```python
df = df.drop(df.loc[df.text.str.fullmatch(r"\s*")].index)
df.shape[0]
```

```python
stopwords = stopwords.words("english") # list stopwords

def cleaner(x):
    
    """This function, when applied to a series, remove figures, 
    words that start with #, @, and https, punctuation, 
    numbers, and stopwords."""
    
    x = x.encode("ascii", "ignore").decode("ascii")
    x = " ".join(i.lower() for i in x.split() if not i.startswith(("#","@","https")))
    x = " ".join(i for i in ["".join(i for i in j.split() if i.isalnum()) for j in x.split()])
    x = "".join(i for i in x if not i.isdigit())
    x = " ".join(i for i in x.split(" ") if i not in stopwords)
    return x
```

```python
df["text"] = df.text.apply(cleaner)
```

```python
df.loc[df.text.str.fullmatch(r"\s*")].head(3)
```

```python
df = df.drop(df.loc[df.text.str.fullmatch(r"\s*")].index)
df.shape[0]
```

```python
df = df.drop_duplicates(subset="text")
df.shape[0]
```

### Word clouds by video + Tf-Idf

```python
by_video = []
for vid in df.video.unique():
    x = df.loc[df.video == vid]
    by_video.append(" ".join(i for i in x.text))  
    
df2 = pd.DataFrame()
df2["video"] = [i for i in df.video.unique()]
df2["text"] = by_video
df2.head()
```

```python
df2.shape[0]
```

```python
vectorizer = TfidfVectorizer(max_df = .6, min_df = .01)
X = vectorizer.fit_transform(df2.text)
feature_names = vectorizer.get_feature_names()
dense = X.todense()
denselist = dense.tolist()
tfidf = pd.DataFrame(denselist, columns=feature_names)
tfidf.head()
```

```python
data = tfidf.transpose()
rndm = np.random.randint(92, size=10)
```

Random numbers:  
85, 19, 56, 12, 20, 7, 35, 37, 51, 14

```python
rndm = [85, 19, 56, 12, 20, 7, 35, 37, 51, 14]
```

```python
N = 1
for i in rndm:
    wordcloud = WordCloud(width=800, height=400).generate_from_frequencies(data[i])
    plt.figure( figsize=(14,7) )
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    wordcloud.to_file("./clouds/image"+str(N)+".png")
    N += 1
    plt.show("png")
```
